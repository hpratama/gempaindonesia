```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.dates as mdates
```

```python
#read data from csv
df = pd.read_csv('All (2).csv', parse_dates=['time'])
#set date as index
df.set_index('time')
df.tail()
```

```python
%matplotlib inline
#Plot depth vs mag
x = df['time']
y = df['mag']
fig, ax = plt.subplots(figsize=(15,7))
plt.plot(x,y,'ro')
plt.show()
```

```python
gempa_besar = df[df.mag >= 6]
gempa_besar.tail()
```

```python
%matplotlib inline
y = gempa_besar['mag']
x = gempa_besar['time']
fig, ax = plt.subplots(figsize=(15,7))
plt.plot(x,y)
plt.show()
```
